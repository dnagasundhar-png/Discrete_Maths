Here is a little sample of plotting timings for a function in Python.

````python
# this first line is just used to make the plots appear nicely in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def fake_function(n):
    time.sleep(0.001*n*n)

def timeit(n):
    start = time.time()
    fake_function(n)
    end = time.time()
    return end-start

N = [0, 10, 20, 30, 40, 50]
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

## Implementation notes on compiled languages

For the other ones, if you're using a compiled language, they might run so fast that in order to time them you should slow down the addition by adding a 1 microsecond wait. I've included a C++ ``busy_sleep`` function below to do that.

```c++
#include <chrono>
#include <iostream>
#include <fstream>

using namespace std;

// this function busy sleeps for n microseconds
void busy_sleep(int n)
{
    auto start = chrono::steady_clock::now();
    while(chrono::duration_cast<chrono::microseconds>(chrono::steady_clock::now() - start).count()<n) {};
}
```

Recursive Method:

````python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def recursive(n):
    if n <= 2:
      return 1

    else:
      return recursive(n-1) + recursive (n-2)


def timeit(n):
    start = time.perf_counter()
    recursive(n)
    end = time.perf_counter()
    return end-start

N = [i for i in range(30)]
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

Table method:

````python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def table(n):
    table = [1,1]
    for i in range(n-2):
        table.append(table[len(table)-1]+table[len(table)-2])
    return table[len(table)-1]


def timeit(n):
    start = time.perf_counter()
    table(n)
    end = time.perf_counter()
    return end-start

N = [i for i in range(100)]
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````

Keep last 2 values:

````python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import time

def last2val(n):
    val_a = val_b = 1

    for i in range(n-2):
        val_a,val_b = val_b, val_a+val_b

    return val_b


def timeit(n):
    start = time.perf_counter()
    last2val(n)
    end = time.perf_counter()
    return end-start

N = [i for i in range(100)]
T = [timeit(n) for n in N]
plt.plot(N, T)
plt.xlabel('n')
plt.ylabel('Time (s)');
````
